## Import Libraries and Load Dataset 

```python
import numpy as np
import pandas as pd
import os

path = "../data/student_scores.csv"
df = pd.read_csv(path)
```

## Convert to NumPy Arrays

```python
scores = df.iloc[:,1:].values
names = df["Name"].values
print(scores)
```

## Student Average Score

```python
student_avg = np.mean(scores, axis=1)

for name, avg in zip(names, student_avg):
    print(name, avg)
```

## Top Performer

```python
top_index = np.argmax(student_avg)

print("Top Student:", names[top_index])
print("Average Score:", student_avg[top_index])
```

## Subject-wise Average

```python
subject_avg = np.mean(scores, axis=0)

subjects = ["Math","Science","English","History"]

for sub, avg in zip(subjects, subject_avg):
    print(sub, avg)
```

## Hardest Subject

```python
hardest = subjects[np.argmin(subject_avg)]

print("Hardest Subject:", hardest)
```

## Pass / Fail Analysis

```python
pass_mask = scores >= 50

pass_count = np.sum(pass_mask, axis=1)

print(pass_count)

fail_students = names[pass_count < 4]

print("Students needing improvement:", fail_students)
```

## Student Ranking

```python
rank = np.argsort(-student_avg)

print("Student Ranking")

for i in rank:
    print(names[i], student_avg[i])
```

## Grade Calculation (Vectorization)

```python
grades = np.where(student_avg >= 90, "A",
         np.where(student_avg >= 80, "B",
         np.where(student_avg >= 70, "C",
         np.where(student_avg >= 60, "D", "F"))))

for name, grade in zip(names, grades):
    print(name, grade)
```
